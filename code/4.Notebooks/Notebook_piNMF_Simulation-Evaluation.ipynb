```python
def percent_reduction(array):
    reduction_percentages = []
    for i in range(len(array) - 1):
        reduction = ((array[i + 1, 3] - array[i, 3]) / array[i, 3]) * 100
        reduction_percentages.append(reduction)
    return reduction_percentages

stats = np.load('/home/jovyan/jm_jlab/jm_2023/time-test/time/time.k_selection_stats.df.npz', allow_pickle=True)

# Calculate reduction percentages
reduction_percentages = percent_reduction(stats['data'])

# Display the results
for i, reduction in enumerate(reduction_percentages):
    print(f"Reduction percentage between k = {i+2} and k = {i+3}: {reduction:.2f}%")
```

```python
df = pd.DataFrame({'Reduction Percentage': reduction_percentages})
```
